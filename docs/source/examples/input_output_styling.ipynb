# Input and Output Styling Examples

This notebook demonstrates how to style input and output areas separately.

```
# Input area with blue accent
def calculate_sum(a, b):
    return a + b

result = calculate_sum(10, 20)
print(f'The sum is: {result}')
```

```
# Output area with green accent
print('=' * 50)
print('OUTPUT STYLING EXAMPLE')
print('=' * 50)
print('This output has custom styling applied')
for i in range(1, 4):
    print(f'Item {i}: Value {i * 10}')
```

```
# Both input and output styled
import random

print('Generating random numbers:')
numbers = [random.randint(1, 100) for _ in range(5)]
print(f'Numbers: {numbers}')
print(f'Average: {sum(numbers) / len(numbers):.2f}')
```

```
# Styled code execution
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print('Fibonacci sequence:')
for i in range(10):
    print(f'F({i}) = {fibonacci(i)}')
```
