# Custom CSS Classes Example

This notebook demonstrates using custom CSS classes with an external stylesheet.

## Important Notice

This cell uses custom CSS classes defined in an external stylesheet.

```
# Code with custom classes
def greet(name):
    return f'Hello, {name}!'

message = greet('World')
print(message)
```

```
# Mixing classes and styles
data = [1, 2, 3, 4, 5]
squared = [x**2 for x in data]
print(f'Original: {data}')
print(f'Squared: {squared}')
```
